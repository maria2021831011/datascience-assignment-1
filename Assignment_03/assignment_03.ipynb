1.Inventory price calculator

```python
def calculate_inventory_summary(prices):
    total_cost=sum(prices)            
    average_cost=total_cost/len(prices)if prices else 0  
    return total_cost,average_cost
user_input=input("Enter product: ")
prices=[int(p.strip())for p in user_input.split(",")]
print(calculate_inventory_summary(prices))


```

2.UNIQUE visitor tracker

```python
def count_unique_visitors(visitor_list):
    unique_visitors=[]
    for vid in visitor_list:
        if vid not in unique_visitors:
            unique_visitors.append(vid)
    return len(unique_visitors),unique_visitors
user_input=input("Enter visitor IDs")
visitor_list=[int(x.strip()) for x in user_input.split(',')]
count,unique_ids = count_unique_visitors(visitor_list)
print(f"{count} (The unique IDs are {', '.join(map(str, unique_ids))})")
```

3.product category counter

```python
def count_categories(products):
    category_count={}
    for product in products:
        if product in category_count:
            category_count[product]+=1
        else:
            category_count[product]=1
    return category_count
user_input=input("Enter product names ")
products=[x.strip() for x in user_input.split(',')]
category_counts=count_categories(products)
print(category_counts)
```

4. Temperature Analytics

```python
def analyze_temperatures(temp_list):
    high_list=[]
    normal_list=[]
    for temp in temp_list:
        if temp > 38.0:
            high_list.append(temp)
        else:
            normal_list.append(temp)
    return high_list,normal_list
user_input=input("Enter temperatures ")
temp_list=[float(x.strip()) for x in user_input.split(',')]
high_temps, normal_temps = analyze_temperatures(temp_list)
print(f"High List (>38째C): {high_temps}")
print(f"Normal List (<38째C): {normal_temps}")
print(f"Full Output Tuple: ({high_temps}, {normal_temps})")
```

5. Menu Price Lookup

```python
def get_price(menu,item):
    return menu.get(item,"Item not found")
menu={
    "Coffee": 3.50,
    "Tea": 2.75,
    "Muffin": 4.00,
    "Sandwich": 8.50
}

item=input("Enter the item name to lookup: ").strip()
price=get_price(menu, item)
print(f"{price:.2f}")
```

6. Student Grading System

```python
def calculate_grade(score):
    if score >= 80:
        return "A+"
    elif score >= 70:
        return "A"
    elif score >= 60:
        return "A-"
    else:
        return "F"

def grade_students(student_dict):
    grades_dict={student: calculate_grade(score) for student, score in student_dict.items()}
    return grades_dict
user_input=input("Enter students and score ")
student_dict={}
for entry in user_input.split(','):
    name,score=entry.split(':')
    student_dict[name.strip()]=int(score.strip())
grades=grade_students(student_dict)
print(grades)
```

7. Discount Calculator

```python
def apply_discounts(products):
    discounted_prices = []
    for name, price, discount in products:
        new_price = price * (1 - discount / 100)
        discounted_prices.append(round(new_price, 2))
    return discounted_prices
user_input = input('Enter products  ')
parts = user_input.split("),")

products = []
for p in parts:
    p = p.strip()
    if not p.endswith(")"):
        p = p + ")"       
    inner = p[1:-1]
    name, price, discount = inner.split(',')
    products.append((name.strip().strip('"').strip("'"), float(price), float(discount)))
result = apply_discounts(products)
formatted = "[" + ", ".join(f"{x:.2f}" for x in result) + "]"
print(formatted)
```

8. Hashtag Generator

```python
def generate_hashtags(sentence):
    words=sentence.split()
    hashtags=['#' + word.lower() for word in words]
    return hashtags

sentence=input("Enter a sentence")
hashtags=generate_hashtags(sentence)
print(hashtags)
```

10. Shopping Cart Bill Calculator

```python
def calculate_total(cart, prices):
    total=0
    for item in cart:
        if item in prices:
            total +=prices[item]
        else:
            print(f"Warning: {item} not found in price list, skipped.")
    return total
cart_input=input("Enter items ")
cart=[item.strip() for item in cart_input.split(',')]
price_input = input("Enter item prices as item ")
prices = {}
for entry in price_input.split(','):
    name, price = entry.split(':')
    prices[name.strip()] = float(price.strip())
total_bill=calculate_total(cart, prices)
print(f"{total_bill:.2f}")
```
