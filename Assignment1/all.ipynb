Temperature Check

```python
a=float(input("enter current temperature celcius"))
if a<10:
    print("It's too cold!")
elif a>=10 and a<=25:  
    print("It's cool outside")
else:
    print("It's hot outside!")      
```

ATM Withdrawal System

```python

account_balance=float(input("enter your balance"))
withdrawal_amount=float(input("enter withdrawal ammount"))
if withdrawal_amount> account_balance:
    print("Insufficient Balance")
else:
    account_balance-=withdrawal_amount
    print("Transaction Successful")
    print("Remaining Balance: ",account_balance)
```

Student Grade Calculator


```python

marks=float(input("enter marks "))
if 90<=marks and marks<=100:
    print("A+")
elif 80<=marks and marks<90:
    print("A")
elif 70<=marks  and marks<80:
    print("B")
elif 60<=marks and marks<70:
    print("C")
else:
    print("Fail")
```

Bus Ticket Discount

```python

age=int(input("enter your age:"))
if age<5:
    print("Ticket Price: Free")
elif 5<=age and age<=18:
    print("Ticket Price: 50% Discount")
elif age>=60:
    print("Ticket Price: 30% Discount")
else:
    print("Ticket Price: Full Price")
```

Restaurant Bill with Discount

```python

bill_amount=float(input("Enter the total bill amounT"))
if bill_amount>1000:
    discount=bill_amount*0.10
    final_bill=bill_amount-discount
    print("Discount applied: 10'%' Final bill amount:",final_bill)
else:
    print("No discount appliedFinal bill amount:",bill_amount)
```

Traffic Light System

```python

color=input("Enter(red, yellow, or green): ").lower()
if color=="red":
    print("Stop")
elif color=="yellow":
    print("Ready to go")
elif color=="green":
    print("Go")
else:
    print("Invalid color")
```

Leap Year Checker

```python

year=int(input("Enter a year: "))
if year%400==0:
    print("Leap Year")
elif year%4==0 and year%100!=0:
    print("Leap Year")
else:
    print("Not a Leap Year")
```
