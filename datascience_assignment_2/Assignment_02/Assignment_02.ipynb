1. Email Validator (String Filtering + Loops)

```python
email=input("Enter your mail:")
flag=False
for i in email:
     if i=="@" :
      flag=True
      break
if flag and (email.endswith(".com")or email.endswith(".org")):
      print("Valid Email")
else:
      print("Inalid Email")          
```

Password Strength Checker

```python
password=input("Enter password:")
up=any(i.isupper() for i in password)
lo=any(i.islower() for i in password)
dg=any(i.isdigit() for i in password)
sp=any(not i.isalnum() for i in password)
if up and lo and dg and sp:
    print("Strong Password")
else:
    print("Weak Password")    
```

Username Generator

```python
import random

full_name = input("Enter full name: ")

fname = ""
lname = ""
flag = False   


for i in full_name:
    if i == " ":
        flag = True
        continue
    if flag == False:
        fname += i
    else:
        lname += i

fast3 = ""
l = 0
for i in fname:
    if l < 3:
        fast3 += i
        l += 1


last3 = ""
length = 0
for i in lname:      
    length += 1

pos = 0
for i in lname:
    if pos >= length - 3:
        last3 += i
    pos += 1
num = ""
for i in range(2):
    num += str(random.randint(0, 9))

print((fast3 + last3 + num).lower())

```

7. Detect Suspicious Words in Text (Moderation)

```python
l=['spam', 'scam', 'fake']
users=input("Eenter sentence").lower()
user=users.split()
war=False
for word in user:
    if word in l:
        war=True
        break
if war:
    print("Warning: Suspicious content detected")
else:
    print("Text is clean")    
```

Count Hashtags and Mentions (Social Media Analytics)

```python
post=input("Enter your post: ")
hashtags=0
mentions=0
for ch in post:
    if ch=='#':
        hashtags +=1
    elif ch=='@':
        mentions+=1
print("Hashtags =",hashtags)
print("Mentions =",mentions)
```

Word Reversal for Data Encryption

```python
sentence=input("Enter a sentence: ")
words=sentence.split()
reversed_words = []

for w in words:
    reversed_words.append(w[::-1])

result = " ".join(reversed_words)
print(result)
```
